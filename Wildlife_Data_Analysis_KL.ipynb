```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#Read in weather data CSV file and drop unnecessary columns
wildlife_df = pd.read_csv("Resources/WildlifeFinal.csv")
wildlife_df = wildlife_df.drop("Unnamed: 0.2", axis='columns')
wildlife_df = wildlife_df.drop("Unnamed: 0.1", axis='columns')
wildlife_df = wildlife_df.drop("Unnamed: 0", axis='columns')
# formatting date 
#pd.to_datetime(wildlife_df["Date"])
wildlife_df['Date'] = pd.to_datetime(wildlife_df['Date'])
wildlife_df.head() 
#view data type of each column in DataFrame
#print(wildlife_df.dtypes)
```

```python
#Read wildlife data csv file
dataset_df = pd.read_csv("Resources/Full-Dataset-(2018-2020).csv")
dataset_df = dataset_df.rename(columns={'Photo.Date': 'Date'})
dataset_df['Date'] = pd.to_datetime(dataset_df['Date'])
dataset_df.head()
#print(dataset_df.dtypes)
```

```python
#Merge DataFrames: Outer Join
wildlife_dfm = pd.merge(wildlife_df, dataset_df, on="Date", how="left")
wildlife_dfm                        
```

```python
wildlife_dfm.to_csv("Resources/wildlife_dfm.csv")
```
