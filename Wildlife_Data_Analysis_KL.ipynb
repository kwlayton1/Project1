```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#Read in weather data CSV file and drop unnecessary columns
wildlife_df = pd.read_csv("Resources/WildlifeFinal.csv")
wildlife_df = wildlife_df.drop("Unnamed: 0.2", axis='columns')
wildlife_df = wildlife_df.drop("Unnamed: 0.1", axis='columns')
wildlife_df = wildlife_df.drop("Unnamed: 0", axis='columns')
# formatting date 
#pd.to_datetime(wildlife_df["Date"])
wildlife_df['Date'] = pd.to_datetime(wildlife_df['Date'])
wildlife_df.head() 
#view data type of each column in DataFrame
#print(wildlife_df.dtypes)
```

```python
#Read wildlife data csv file
dataset_df = pd.read_csv("Resources/Full-Dataset-(2018-2020).csv")
dataset_df = dataset_df.rename(columns={'Photo.Date': 'Date'})
dataset_df['Date'] = pd.to_datetime(dataset_df['Date'])
dataset_df.head()
#print(dataset_df.dtypes)
```

```python
#Merge DataFrames: Left Join
wildlife_dfm = pd.merge(wildlife_df, dataset_df, on="Date", how="left")
wildlife_dfm.head()                        
```

```python
#exported wildlife_dfm to csv to share with the whole group
#wildlife_dfm.to_csv("Resources/wildlife_dfm.csv")
#to view animal and number of sightings
animal_count = wildlife_dfm["Common.Name"].value_counts()
animal_count
```

```python
# Specify the column name and the list of desired values you want to filter on
column_name = 'Common.Name'
desired_values = ['Human non-staff', 'Bicycle', 'Mule Deer', 'Northern Raccoon', 'Red Fox', 'Striped Skunk',  'Moose', 'Coyote', 'Elk aka Red Deer', 'Puma', 'American Beaver', 'Bobcat']

# Create a new DataFrame by filtering the rows based on the desired values using isin method
human_wildlife_df = wildlife_dfm[wildlife_dfm[column_name].isin(desired_values)].copy()

# If you want to reset the index of the new DataFrame
human_wildlife_df.reset_index(drop=True, inplace=True)
human_wildlife_df

#Get only 2018 data
# Specify the column name that contains the dates
date_column = 'Date'

# Convert the column to pandas datetime if it's not already in datetime format
human_wildlife_df['Date'] = pd.to_datetime(human_wildlife_df['Date'])

# Create a new DataFrame by filtering the rows based on the year 2018
human_wildlife_2018 = human_wildlife_df[human_wildlife_df['Date'].dt.year == 2018].copy()

# If you want to reset the index of the new DataFrame
human_wildlife_2018.reset_index(drop=True, inplace=True)
human_wildlife_2018

#Get only 2019 data
# Specify the column name that contains the dates
date_column = 'Date'

# Convert the column to pandas datetime if it's not already in datetime format
human_wildlife_df['Date'] = pd.to_datetime(human_wildlife_df['Date'])

# Create a new DataFrame by filtering the rows based on the year 2019
human_wildlife_2019 = human_wildlife_df[human_wildlife_df['Date'].dt.year == 2019].copy()

# If you want to reset the index of the new DataFrame
human_wildlife_2019.reset_index(drop=True, inplace=True)
human_wildlife_2019

#Get only 2020 data
# Specify the column name that contains the dates
date_column = 'Date'

# Convert the column to pandas datetime if it's not already in datetime format
human_wildlife_df['Date'] = pd.to_datetime(human_wildlife_df['Date'])

# Create a new DataFrame by filtering the rows based on the year 2020
human_wildlife_2020 = human_wildlife_df[human_wildlife_df['Date'].dt.year == 2020].copy()

# If you want to reset the index of the new DataFrame
human_wildlife_2020.reset_index(drop=True, inplace=True)
human_wildlife_2020
```
